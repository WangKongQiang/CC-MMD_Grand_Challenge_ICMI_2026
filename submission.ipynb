models--mlabonne--gemma-3-1b-it-abliterated

```python
import json
import pandas as pd

# ===== 文件路径 =====
india_path = "Gemma-3-1B-it-abliterated/results_test_Tamil_India/models--mlabonne--gemma-3-1b-it-abliterated_tamil.json"
ireland_path = "Gemma-3-1B-it-abliterated/results_test_Tamil_Ireland/models--mlabonne--gemma-3-1b-it-abliterated_tamil.json"
china_path = "Gemma-3-1B-it-abliterated/results_test_Tamil_China/models--mlabonne--gemma-3-1b-it-abliterated_tamil.json"

# ===== 读取 JSON =====
def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

india_data = load_json(india_path)
ireland_data = load_json(ireland_path)
china_data = load_json(china_path)

# ===== 转换为 DataFrame =====
def to_df(data, col_name):
    df = pd.DataFrame(data)
    
    # label 映射
    df[col_name] = df["classification"].map({
        "misogyny": 1,
        "not-misogyny": 0
    })
    
    return df[["image_id", col_name]]

df_india = to_df(india_data, "original_culture")
df_ireland = to_df(ireland_data, "irish_culture")
df_china = to_df(china_data, "chinese_culture")

# ===== 合并（按 image_id 对齐）=====
df = df_india.merge(df_ireland, on="image_id")
df = df.merge(df_china, on="image_id")

# ===== 排序（可选但推荐）=====
# df = df.sort_values(by="image_id")

# ===== 保存 =====
# output_path = "Gemma-3-1B-it-abliterated/tamil/teamname_taskb_tamil.csv"
output_path = "Gemma-3-1B-it-abliterated/tamil/wangkongqiang_taskb_tamil.csv"
df.to_csv(output_path, index=False)

print("✅ Saved to:", output_path)
print(df.head())
```

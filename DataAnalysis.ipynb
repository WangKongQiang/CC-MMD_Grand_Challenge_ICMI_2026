帮我书写一个python脚本能够统计train.csv的transcriptions的文本长度情况，包括平均值，最大值，最小值。以及original_labels，irish_labels, chinese_labels的标签数量情况。

```python
import pandas as pd

# 读取 CSV 文件
df = pd.read_csv("train/train.csv")

# =========================
# 1. transcription 文本长度统计
# =========================

# 防止空值影响
df["transcriptions"] = df["transcriptions"].fillna("").astype(str)

# 计算文本长度
df["text_length"] = df["transcriptions"].apply(len)

# 统计结果
avg_length = df["text_length"].mean()
max_length = df["text_length"].max()
min_length = df["text_length"].min()

print("===== Transcriptions 文本长度统计 =====")
print(f"平均长度: {avg_length:.2f}")
print(f"最大长度: {max_length}")
print(f"最小长度: {min_length}")

# =========================
# 2. 标签数量统计
# =========================

label_columns = [
    "original_labels",
    "irish_labels",
    "chinese_labels"
]

for col in label_columns:
    print(f"\n===== {col} 标签统计 =====")
    
    # 统计每个标签数量
    label_count = df[col].value_counts().sort_index()
    
    print(label_count)
```
